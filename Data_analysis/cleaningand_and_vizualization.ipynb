```python
import pandas as pd
```

```python
df = pd.read_csv('laptops_cleaned.csv')
```

```python
df.info()
```

```python
# G-1 Higher-priced laptops tend to receive higher customer ratings. 
higher_price = df['laptop-price'].mean()
higher_rating = df['laptop-rating'].mean()
result_price = df[df['laptop-price'] >= higher_price]
finally_result = result_price.set_index('laptop-brand')[['laptop-price', 'laptop-rating']]
import matplotlib.pyplot as plt
finally_result.plot.scatter(
    title = '# G-1 Higher-priced laptops tend to receive higher customer ratings.',
    x='laptop-price',
    y='laptop-rating',
    figsize=(10,6),
    color='blue',
    s=100
)
```

```python
# G-2 Premium brands like HP and Dell have higher average ratings than lesser-known brands.
all_brand_rating = df.groupby(by='laptop-brand')['laptop-rating'].mean()
ax= all_brand_rating.plot(kind='barh',
                         figsize=(5,5),
                          title="# G-2 Premium brands like HP and Dell have higher average ratings than lesser-known brands.")
ax.set_xticks([0,1,1.5,2,2.5,3,3.5,4,4.5,5])
plt.show()
```

```python
# G-3 Laptops with larger screens (15 inches and above) receive better customer ratings compared to smaller screens (14 inches and below).
df.set_index('laptop-brand')
res1 = df[df['laptop-screensize'] >= 15]
high_screen = res1.set_index('laptop-brand')[['laptop-screensize','laptop-review_count']]
res2 = df[df['laptop-screensize'] < 15]
lower_screen = res2.set_index('laptop-brand')[['laptop-screensize','laptop-review_count']]
high_avg_review = high_screen['laptop-review_count'].mean()
lower_avg_review = lower_screen['laptop-review_count'].mean()


x_labes = ['High screen', 'Lower screensize']
values = [high_avg_review, lower_avg_review]
plt.bar(x_labes,values, color=['red', 'blue'])
plt.ylabel('Avg review count')
plt.title('# G-3 Laptops with larger screens (15 inches and above) receive better customer ratings compared to smaller screens (14 inches and below)')
plt.show()
```

```python
# G-4 Laptops running Windows 11 are rated higher than those using older versions (Windows 10)
total_rating = df[['laptop-operating_system','laptop-rating']]
mean_rating = total_rating[total_rating['laptop-operating_system'] == 'Windows 11']['laptop-rating'].mean()
win_10_rating = total_rating[(total_rating['laptop-operating_system'] == 'Windows 10 Home') | (total_rating['laptop-operating_system'] == 'Windows 10 Pro')]['laptop-rating'].mean()
plt.figure(figsize=(6,4))
plt.barh(['Windows 11','Windows 10'], [mean_rating,win_10_rating], color=['red', 'blue'])
plt.xlabel('Avg rating')
plt.ylabel('OS system')
plt.title('# G-4 Laptops running Windows 11 are rated higher than those using older versions (Windows 10)')
plt.show()
```

```python
# G-5 Laptops with longer customer reviews tend to have higher overall ratings
mean_rating = df['laptop-rating'].mean()
mean_review = df['laptop-review_count'].mean()

high_review = df.set_index('laptop-brand')[['laptop-rating','laptop-review_count']]
total1 = high_review[high_review['laptop-review_count'] >= mean_review]
all = total1['laptop-rating'].count()
son1 = total1[total1['laptop-rating'] >= mean_rating]
all_high_rating = son1['laptop-rating'].count()
all_lower = all - all_high_rating

labels = 'high review and high rating','high review and lower rating'
sizes = [all_high_rating,all_lower]
fig,ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
# G-6 Mid-range laptops ($500â€“$800) receive better ratings than budget or premium models.
# G-7 Higher laptops rating tand to have higher laptops storage
# G-8 Higher laptops review tand to have higher laptops ram
```

```python
# G-6 Mid-range laptops ($500â€“$800) receive better ratings than budget or premium models.
result_pr = df.set_index('laptop-brand')[['laptop-price','laptop-rating']]
max_price = result_pr['laptop-price'].max()
min_price = result_pr['laptop-price'].min()
rating1 = result_pr[(result_pr['laptop-price'] >= 500) & (result_pr['laptop-price'] <= 800)]
ras_price = rating1['laptop-price'].mean()
ras_rating = rating1['laptop-rating'].mean()
max_rating = result_pr[result_pr['laptop-price'] == max_price]
max_ratingnew = max_rating['laptop-rating'].sum()
min_price = result_pr[result_pr['laptop-price'] == min_price]
min_ratingnew = min_price['laptop-rating'].sum()

fig,ax = plt.subplots(figsize=(8,4))

ax.barh(['Min price', '500$-800$ Price', 'Max price'], [min_ratingnew,ras_rating,max_ratingnew], color=['red','green','blue'])
ax.set_xticks([0,1,1.5,2,2.5,3,3.5,4,4.5,5])

ax.set_yticks([0,1,2])
ax.set_title('Min price and (500$-800$ Price) and Max price')
plt.show()
```

```python
# G-7 Higher laptops rating tand to have higher laptops storage
total_sr = df.set_index('laptop-brand')[['laptop-storage','laptop-rating']]
high_rating = total_sr['laptop-rating'].mean()
high_storage = total_sr['laptop-storage'].median()
res1 = total_sr[total_sr['laptop-storage'] >= high_storage]
all_storage = res1['laptop-rating'].count()
higher_ratings = res1[res1['laptop-rating'] >= high_rating]['laptop-rating'].count()
lower_ratings = all_storage - higher_ratings

labels = 'high storage and high rating','high storage and lower rating'
sizes = [higher_ratings,lower_ratings]
fig,ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
# G-8 Higher laptops review tand to have higher laptops ram
mean_review = df['laptop-review_count'].mean()
mean_ram = df['laptop-ram'].median()
rev_ram = df.set_index('laptop-brand')[['laptop-ram','laptop-review_count']]
result_rr = rev_ram[rev_ram['laptop-ram'] >= mean_ram]

all_rr = result_rr.groupby(by='laptop-ram')['laptop-review_count'].count()
kk = all_rr.plot(
    kind='barh',
    figsize=(4,4),
    title='# G-8 Higher laptops review tand to have higher laptops ram',
    color=['lightcoral', 'lightgreen', 'lightblue', 'orange', 'purple']
)
kk.set_xticks([0,10,20,30,40,50])
plt.show()
```
